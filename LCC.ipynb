## Modification of LCC current voltage profile to resemble -Rad data

### Questions:
- (fCa) f in LCC model refer to the fraction of (calcium-dependent) inactivation gates, yet LCC current is given by

$ I = d\cdot fCa\cdot f\cdot Q_{10} I_{GHK} $

where $d$ is the fraction of activation gates and $I_{GHK}$ is the Goldman Hodgkin Katz flux model. Apparently f>0 refers to an open state, as f converges from a positive value to 0 as t increases

### ODE simulation for LCC channel over voltage range

```python
Vs = np.linspace(-90,80,35)
#print Vs
Vs+=0.001
```

```python
import runner_LCC 

runner_LCC.init()
baselineParams = np.copy(runner_LCC.model.p)
param_indices = runner_LCC.model.param_indices

printRate = 5
def CalcLCC(ps=None,tag="_allion"):
  peaks = []
  rVs=[]
  for i,V in enumerate(Vs):
    (p,si,ts,rV) = runner_LCC.runner(dt=100,dtn=0.05,V=V, ps=ps)
    if i % 10 == -1: # to see the d/f distros
      runner_LCC.plotting(p,si,ts,rV)
      plt.title("V=%3.0f"%V)
    rVs.append(rV)
    maxrV = np.min(rV)
    peaks.append(maxrV)
  #return maxrV,peaks,ts  
  
  Plotter(rVs,peaks,ts,mode="IV",tag=tag)    
  return rVs,peaks  
    
```

Create interpolation function and estimate I(V(t))

### Current wrt time and total Ca2+ moved. 
Main conclusion: PCa has the largest effect compared to shift in voltage. 
- Note I here looks a lot different than action potential in complete ode model, which may be because Cai changes in the Shannon-Bers model, while here Cai is held constant. 

Using

$\phi [uM/ms] = 1/2F_c [umol/C] \cdot \frac{C_m}{V_m} [F/L] \cdot I [A/F] \approx I\cdot4.5\cdot 5.2\cdot 10^{-3} [uM/ms]$

and

$\phi  \approx I\cdot 2\cdot 10^{-2} [uM/ms]$

Need to double check the I to phi conversion once more, but note that Ca brought in via LCC is quickly buffered by TnC, which greatly suppressed the increase in free cytosolic Ca


Create interpolation function and estimate I(V(t))

```python
from scipy.interpolate import interp1d
IphiConv=2e-2 # [A/F]*[uM/ms]
def interps(Vs,peaks,Vt):
  func = interp1d(Vs,peaks,kind='cubic')
  IV = func(Vt)
  dt = ts[1]-ts[0]
  IVdT = -1*IV*dt*IphiConv
  return IV,IVdT
```

```python
## I(t)
plt.figure()
plt.title("I(t) over action potential")
for i,peaksi in enumerate(allPeaks):
 IV,phi = interps(Vs,peaksi,Vt)   
 plot(ts,-1*IV, cols[i],label=labels[i])

plt.xlabel("t [ms]")
plt.ylabel("I(t) [A/F]")
plt.legend(loc=0)    
plt.gcf().savefig("LCC_over_AP.png")


## I(t)
plt.figure()
plt.title("Integrated $\phi(t)\cdot dt$ [uM] over action potential")
for i,peaksi in enumerate(allPeaks):
 IV,phi = interps(Vs,peaksi,Vt)   
 plot(ts,np.cumsum(phi), cols[i],label=labels[i])

plt.xlabel("t [ms]")
#plt.ylabel("I(t)*dt [A*s/F]")
plt.ylabel("$Integrated \phi(t)*dt $ [uM]")
plt.legend(loc=0)    
plt.gcf().savefig("fCai_over_AP.png")
```

```python
import shannon_2004 as modelimport separatefrom scipy.integrate import odeintCai_idx =  model.state_indices( "Cai" )V_idx =  model.state_indices( "V" )stim_period_pIdx =  model.param_indices( "stim_period" )Kmf_pIdx =  model.param_indices( "Kmf" )class empty:passdef dostuff(Kmf_in=Kmf,dt=1000,stim_period=1000,model_in=model):  t=0  dtn=5  mxsteps=500  s=model_in.init_values()  p=model_in.default_parameters()      p[stim_period_pIdx] = stim_period  p[Kmf_pIdx] = Kmf_in  tsteps = np.linspace(t, t+dt, (dt)/dtn+1)  states = odeint(model_in.rhs,s,tsteps,(p,),mxstep=mxsteps)      jall = np.zeros(tsteps.shape)  for i,t in enumerate(tsteps):    # get current state    si = states[i,:]    # extract monitored fluxes     jis = model_in.monitor(si, t, p)    #print np.shape(jis)    jall[i] = jis[separate.j_pump_SR_idx]  results = empty()  results.tsteps = tsteps  results.states = states  results.js = jall    return (results)
```

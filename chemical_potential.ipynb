```python
b=23. # [nm]
V=b**3
k=0.3*b
u_0_rep=5.*300*1.3806*10**-23 #*kT
u_0_att=-5.*300*1.3806*10**-23 #*kT
k
```

```python
u_e_rep0=u_0_rep*exp(-dist0/k)
u_e_rep0
```

```python
u_e_att0=u_0_att*exp(-dist0/k)
u_e_att0
```

```python
u_e_rep1=u_0_rep*exp(-dist1/k)
u_e_rep1
```

```python
u_e_att1=u_0_att*exp(-dist1/k)
u_e_att1
```

```python
u_e_rep2=u_0_rep*exp(-dist2/k)
u_e_rep2
```

```python
u_e_att=np.array((u_e_att0, u_e_att1, u_e_att2, u_e_att3))
u_tot_att=np.sum(u_e_att)
u_tot_att
```

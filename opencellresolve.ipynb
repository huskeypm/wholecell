### ODE int

```python
t0=0
t1 = 500
dt = 1.
tsteps=linspace(t0,t1,(t1-t0)/dt)

def fodeint(s,t,arg1):
    eq1 = 0.1*arg1*s[0] + 0.1* s[1]
    eq2 = -arg1*s[1]*2    
    return [eq1,eq2]

state0=[10,2]
a=0.02


solnODEint = odeint(fodeint,state0,tsteps,args=(a,))
```

### ODE

```python
from scipy.integrate import ode

def fode(t, s, arg1):
    return fodeint(s,t,arg1)
    #eq1 = 0.1*arg1*s[0] + 0.1* s[1]
    #eq2 = -arg1*s[1]*2    
    #return [eq1,eq2]


r = ode(fode).set_integrator('zvode', method='bdf', with_jacobian=False)
r.set_initial_value(state0, t0).set_f_params(a)

solnODE = np.zeros([np.shape(t)[0],np.shape(state0)[0]])
i=0
while r.successful() and r.t < t1:
   r.integrate(r.t+dt)
   #print("%g %g" % (r.t, r.y[0]))
   solnODE[i,:]=r.y
   i+=1 

    
```

```python
plot(solnODEint[:,0],solnODEint[:,1],'k-')
plot(solnODEint[:,0],solnODEint[:,1],'k.')
plot(solnODE[:,0],solnODE[:,1],'b-')
plot(solnODE[:,0],solnODE[:,1],'b.')
```

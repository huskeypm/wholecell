Read in data 

```python
import cPickle as pickle
import runner 
runner.init()

def readOut(name = "PCa0.75kss0.25.pickle"):
  #print name  
  pkl_file = open(name, 'rb')
  data1 = pickle.load(pkl_file)  
  pkl_file.close()
     
  return data1  
    

def analOut(data1,label=""):
  si = data1['s']
  pi = data1['p']

  np.shape(si)
  temp = si[:,runner.model.state_indices("Cai")]
  # take last third
  idx = np.int(np.shape(temp)[0]*2/3)
  #print idx
  ts = data1['t']
  ts = ts[idx:]
  temp = temp[(1+idx):]
  plot(ts,temp,label=label)
    
  PCa = pi[runner.model.param_indices("PCa")]
  ks = pi[runner.model.param_indices("ks")]
  minCai = np.min(temp)
  maxCai = np.max(temp)  
    
  return (PCa,ks,minCai,maxCai)
    
#name =  "/tmp/PCa1.00ks1.00.pickle"      
#d = readOut(name)
#pca,ks,minCai,maxCai = analOut(d)


    
        
```

```python
outsMin = np.zeros([np.shape(PCas)[0],np.shape(kss)[0]])
outsMax = np.zeros([np.shape(PCas)[0],np.shape(kss)[0]])
for i, PCa in enumerate(PCas):
  plt.figure()  
  plt.title("PCa=%3.2f"%PCa)
  for j, ks in enumerate(kss):
    name =namer(PCa,ks)+".pickle"
    #print name
    d = readOut(name) 
    #print np.shape(d['s'])
    pca,ks,minCai,maxCai = analOut(d,label="ks=%3.2f"%ks)
    outsMin[i,j]= minCai 
    outsMax[i,j]= maxCai 
  plt.ylim([1e-4,1.5e-3])  
  plt.legend(loc=0)  
```

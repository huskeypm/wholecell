Loads results from 'runShannonTest.py' executed via runem, which performs a parameter sweep over two variables controlling LCC or RyR Ca2+ release.

- LCC model:

$ I_{LCC} = d\cdot fCa\cdot f\cdot Q_{10} I_{GHK} $

where $d$ is the fraction of activation gates and $I_{GHK}$ is the Goldman Hodgkin Katz flux model. 

$I_{GHK} = P_{Ca} z^2 V \frac{F^2}{RT} 
\frac{ \gamma_i [Ca_i] exp \left(\frac{zVF}{RT} \right) - \gamma_o [Ca_o]}
{exp\left(\frac{zVF}{RT}\right)-1}$

- RyR:

$J_{SRCaRel} = k_s O ([Ca_{SR}] - [Ca_{jct}])$

where $O$ is the open-state probability

```python
import cPickle as pickle
import runner 
runner.init()

mM_to_uM=1e3

def readOut(name = "PCa0.75kss0.25.pickle"):
  print "Reading " + name  
  pkl_file = open(name, 'rb')
  data1 = pickle.load(pkl_file)  
  pkl_file.close()
     
  return data1  
    

def analyOut(data1,state="Cai",label=""):
  si = data1['s']
  pi = data1['p']

  np.shape(si)
  temp = si[:,runner.model.state_indices(state)]
  # take last third
  idx = np.int(np.shape(temp)[0]*2/3)
  #print idx
  ts = data1['t']
  ts = ts[idx:]
  temp = temp[(1+idx):]
  plot(ts,temp*mM_to_uM,label=label)
    
  PCa = pi[runner.model.parameter_indices("PCa")]
  ks = pi[runner.model.parameter_indices("ks")]
  minCai = np.min(temp)
  maxCai = np.max(temp)  
    
  return (PCa,ks,minCai,maxCai)
    
#name =  "/tmp/PCa1.00ks1.00.pickle"      
#d = readOut(name)
#pca,ks,minCai,maxCai = analyOut(d)


    
        
```

```python
def ProcessOutputs(state="Cai",ylims=None):
  outsMin = np.zeros([np.shape(PCas)[0],np.shape(kss)[0]])
  outsMax = np.zeros([np.shape(PCas)[0],np.shape(kss)[0]])
  for i, PCa in enumerate(PCas):
    plt.figure()  
    plt.title("PCa=%3.2f"%PCa)
    for j, ks in enumerate(kss):
        name =namer(PCa,ks)+".pickle"
        #print name
        d = readOut(name) 
        #print np.shape(d['s'])
        pca,ks,minCai,maxCai = analyOut(d,state=state,label="ks=%3.2f"%ks)
        outsMin[i,j]= minCai 
        outsMax[i,j]= maxCai 
    if ylims!=None:    
      plt.ylim(ylims)  
    plt.ylabel(state+" [uM]")
    plt.xlabel("t [ms]")  
    plt.legend(loc=0,ncol=3)  
    name = state+"transientsPCa%3.2f"%PCa
    plt.gcf().savefig(name.replace(".","p")+".png")
    
  return outsMin,outsMax
```

```python
def TwoDPlots(outsMin, outsMax,state="Cai"):
    xv, yv = meshgrid(kss,PCas)

    plt.figure()
    subplot(2,2,1)
    plt.title(state+" min")
    pcolormesh(xv,yv,outsMin*mM_to_uM)#,shading='gouraud')
    plt.ylabel("PCa [L/(F ms)]")
    plt.colorbar()

    subplot(2,2,2)
    plt.title(state+" max")
    pcolormesh(xv,yv,outsMax*mM_to_uM)
    plt.ylabel("PCa [L/(F ms)]")
    plt.colorbar()
    plt.xlabel("ks [1/ms]")
    
    
    subplot(2,2,3)
    plt.title(state+" diff")
    pcolormesh(xv,yv,(outsMax-outsMin)*mM_to_uM)
    plt.ylabel("PCa [L/(F ms)]")
    plt.colorbar()
    plt.xlabel("ks [1/ms]")    
    plt.tight_layout()
    
    name = state+"_extrema.png"
    plt.gcf().savefig(name)
```

```python
idxCai = runner.model.state_indices("Cai")
idxCaSR = runner.model.state_indices("Ca_SR")
def doprocess(case,pacingInterval,tstart=8000):
    tau = GetTau(case,pacingInterval,tstart=tstart,idxCai=idxCai)
    
    tsub, caisub= GetInterval(case,pacingInterval,tstart=tstart,idx=idxCai)
    delCa = (np.max(caisub) - np.min(caisub))*mM_to_uM
    tsub, casrsub= GetInterval(case,pacingInterval,tstart=tstart,idx=idxCaSR)
    delCaSR = (np.max(casrsub) - np.min(casrsub))*mM_to_uM
    print delCa, delCaSR
    #pctChg = GetExtreme(t,cai,subMin=7900,subMax=8500,si=7910)
    #pctChgSR = GetExtreme(t,casr,subMin=7900,subMax=8500,si=7910)
    
    return tau,delCa, delCaSR


cases=["baseline",\
           "pca1p25","vmax2p00","pca1p25vmax2p00"]
casesNames=["baseline",\
           "LCC+25%","SERCA+100%","LCC+25%, SERCA+100%"]
class empty:pass
mM_to_uM = 1e3
def ProcessAll(pacingInterval,tag="",cols=[],name=None):
    
    baseline = readOut("PCa1.00ks1.00vMax1.00"+tag+".pickle")
    pca1p25 = readOut("PCa1.25ks1.00vMax1.00"+tag+".pickle")
    vmax2p00 = readOut("PCa1.00ks1.00vMax2.00"+tag+".pickle")
    #pca1p25vmax1p25 = readOut("PCa1.25ks1.00vMax1.25"+tag+".pickle")
    pca1p25vmax2p00 = readOut("PCa1.25ks1.00vMax2.00"+tag+".pickle")

    
    taus=[]
    pctChgs=[]
    pctChgSRs=[]
    for i, case in enumerate(cases):
      var = eval(case)
      tau,pctChg,pctChgSR  = doprocess(var,pacingInterval)
      taus.append(tau)
      pctChgs.append(pctChg)
      pctChgSRs.append(pctChgSR)  

    plt.figure()    
    plt.subplot(1,2,1) 
    for i, case in enumerate(cases): 
      var = eval(case)
      s = var['s']
      t = var['t']
      cai = s[1:,idxCai]
      idx = 8000
      plt.plot(t[idx:],cai[idx:]*mM_to_uM,cols[i],label=case)
      plt.xlabel("t [ms]")
      plt.ylabel("Ca [uM]")          
    #plt.legend(loc=3,ncol=2)
    plt.title("Cai")
    plt.ylim([0,1.2]) 
    
    plt.subplot(1,2,2)    
    for i, case in enumerate(cases):
      var = eval(case)  
      s = var['s']
      t = var['t']
      casr = s[1:,idxCaSR]
      idx = 8000
      plt.plot(t[idx:],casr[idx:]*mM_to_uM,cols[i],label=case)
      plt.xlabel("t [ms]")
      plt.ylabel("Ca [uM]")
    plt.ylim([250,1e3])    
    plt.legend(loc=2,ncol=1)
    plt.title("CaSR")
    
    plt.tight_layout()
    plt.gcf().savefig(name,dpi=300)
    
    results = empty()
    results.taus = taus; results.pctChgs = pctChgs; results.pctChgSRs = pctChgSRs
    return results

#cases=["baseline"]
#cases=["pca1p25"]
#cases=["pca1p25vmax1p25"]
#cases=["pca1p25vmax1p75"]


cols = ['b.','b:','b-','b--'] 
pacingInterval = 701
results701 = ProcessAll(pacingInterval,tag="stim701",cols=cols, name="stim701.png")
cols = ['r.','r:','r-','r--'] 
pacingInterval = 411
results441 = ProcessAll(pacingInterval,tag="",cols=cols, name="stim441.png")
```

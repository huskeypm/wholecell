
- LCC model:

$ I_{LCC} = d\cdot fCa\cdot f\cdot Q_{10} I_{GHK} $

where $d$ is the fraction of activation gates and $I_{GHK}$ is the Goldman Hodgkin Katz flux model. 

$I_{GHK} = P_{Ca} z^2 V \frac{F^2}{RT} 
\frac{ \gamma_i [Ca_i] exp \left(\frac{zVF}{RT} \right) - \gamma_o [Ca_o]}
{exp\left(\frac{zVF}{RT}\right)-1}$


Also modified Ca2+ inactivation expression to include forward parameter, $k^+$:

$ fCa = 1 - fCaB$

$\frac{dfCa}{dt} = k^+ [Ca](1-fCaB) - 11.9 fCaB$

- RyR:

$J_{SRCaRel} = k_s O ([Ca_{SR}] - [Ca_{jct}])$

where $O$ is the open-state probability

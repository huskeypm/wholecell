### GotranJIT.py
gotranJIT.py allows you to quickly test your ode model, but won't offer much functionality for changing parameters, etc. '-doit' needs to be the last argument in your command string. 

<code>
source configgotran.bash 
python gotranJIT.py -file_name shannon_2004_mouseIto.ode -stim_period 1000 -doit
</code>

I've verified that the model runs for different pacings (1-2 Hz). If you want to modify anything else, look at the code. 

#### Test w mousified model 

```python
import gotranJIT
file_name = "shannon_2004_mouseIto.ode"
stateName = "V" # Cai
gotranJIT.doit(file_name = file_name,
               stateName=stateName,
               stim_period = 500 ) # [ms]
```

```python
gotranJIT.plotit()
```

### runShannonTest.py

#### Default shannon_2004.ode
python runShannonTest.py -stim 1000 -jit -T 30000 -var PCa 0.001080 -var lccCaInact 5.950000 -name test_incrLCC_1000.pickle &

```python
import analyzeODE as ao
```

```python
data = ao.readPickle("test_incrLCC_1000.pickle")

ms_to_s = 1e-3
si = data['s']
s_idx = data['s_idx']
t = data['t'] * ms_to_s

# ca 
subplot(2,2,1)
label = "Cai"
idx = s_idx.index(label)
plt.plot(t,si[:,idx])
plt.title(label)

subplot(2,2,3)
plt.plot(t,si[:,idx])
plt.xlim([25e3*ms_to_s,30e3*ms_to_s])

# V
subplot(2,2,2)
label = "V"
idx = s_idx.index(label)
plt.plot(t,si[:,idx])
plt.title(label)

subplot(2,2,4)
plt.plot(t,si[:,idx])
plt.xlim([25e3*ms_to_s,30e3*ms_to_s])


```

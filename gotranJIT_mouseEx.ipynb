### GotranJIT.py
gotranJIT.py allows you to quickly test your ode model, but won't offer much functionality for changing parameters, etc. '-doit' needs to be the last argument in your command string. 

<code>
source configgotran.bash 
python gotranJIT.py -file_name shannon_2004_mouseIto.ode -stim_period 1000 -doit
</code>

I've verified that the model runs for different pacings (1-2 Hz). If you want to modify anything else, look at the code. 

#### Test w mousified model 

```python
import gotranJIT
file_name = "shannon_2004_mouseIto.ode"
stateName = "V" # Cai
gotranJIT.doit(file_name = file_name,
               stateName=stateName,
               stim_period = 500 ) # [ms]
```

```python
states = ["V", "Cai", "Nai"]
for i,state in enumerate(states):
    ao.PlotPickleData(data,idxName=state, label1=state,trange=[25e3,25.3e3])

fluxes = ["i_Na", "i_tof", "i_Kr"]
for i,flux in enumerate(fluxes):
    ao.PlotPickleData(data,idxName=flux,label1="%s [A/F]"%flux,trange=[25e3,25.3e3]) 
```

$I_K$ is the sum over several potassium currents, so extra work is needed 

```python
fluxes = ["i_tof", "i_Kr"]  # add the rest of the contributors here 

t = data['t']
iK = np.zeros(np.shape(t)[0])

j = data['j']
j_idx = data['j_idx']
for i,flux in enumerate(fluxes):
    idx1 = j_idx.index(flux)
    iK+=j[:,idx1]
    
plt.plot(t,iK,label="IK [A/F]")
plt.legend()
```

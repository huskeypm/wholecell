Exp diastolic Ca

```python
data = np.loadtxt("data/circ_res_SIV_b.txt",skiprows=3)
ctlDias = data[:,[0,1,2]]
amyDias = data[:,[0,3,4]] # 50 uM, skipping 25 uM
```

```python
ax = plt.subplot()
# Trans
ax.errorbar(ctlTrans[:,0], ctlTrans[:,1], yerr=ctlTrans[:,2],color='k',label="Ctl")
ax.errorbar(amyTrans[:,0], amyTrans[:,1], yerr=amyTrans[:,2],color='b',label="Amylin (50 uM)")
ax.set_ylabel("Amplitude ($\Delta F/F_0$)")
ax.set_xlabel("Frequency [Hz]")
ax.legend(loc=2)
# Dais
ax2 = ax.twinx()
ax2.errorbar(ctlDias[:,0], ctlDias[:,1], yerr=ctlDias[:,2],color='k',fmt='--',label="Ctl")
ax2.errorbar(amyDias[:,0], amyDias[:,1], yerr=amyDias[:,2],color='b',fmt='--',label="Amylin (50 uM)")
ax2.set_ylabel("Diastolic [Ca] ($F_{340}/F_{380}$)")

ax.set_xlim([0,2.125])

plt.gcf().savefig("/net/share/shared/papers/despa/despa_amy.png",dpi=300)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
font = {'weight' : 'bold',
        'size'   : 12}
matplotlib.rc('font', **font)
```

```python
time=np.loadtxt('morotti/wt/time.csv')
print time
V=np.loadtxt('morotti/wt/action_potential.csv')
print V
Cai=np.loadtxt('morotti/wt/Cai.csv')
print Cai
CaSR=np.loadtxt('morotti/wt/CaSR.csv')
print CaSR
```

```python
time_amylin=np.loadtxt('morotti/original/morotti_et_al_mouse_model/time_amylin.csv')
V_amylin=np.loadtxt('morotti/original/morotti_et_al_mouse_model/V_amylin.csv')
Cai_amylin=np.loadtxt('morotti/original/morotti_et_al_mouse_model/Cai_amylin.csv')
CaSR_amylin=np.loadtxt('morotti/original/morotti_et_al_mouse_model/CaSR_amylin.csv')
```

```python
time_hip=np.loadtxt('morotti/original/morotti_et_al_mouse_model/time_hip.csv')
V_hip=np.loadtxt('morotti/original/morotti_et_al_mouse_model/V_hip.csv')
Cai_hip=np.loadtxt('morotti/original/morotti_et_al_mouse_model/Cai_hip.csv')
CaSR_hip=np.loadtxt('morotti/original/morotti_et_al_mouse_model/CaSR_hip.csv')
```

```python
plt.plot(time,V,color="blue",label="Baseline")
plt.plot(time_amylin,V_amylin,color="red",label="Amylin Incubation")
plt.plot(time_hip,V_hip,color="black",label="+HIP")
plt.legend(bbox_to_anchor=(1.75,1.0),fancybox=True, shadow=True)
plt.xlim([58.5,60])
plt.ylabel("V [V]",**font)
plt.xlabel("Time [s]",**font)  
```

```python
plt.plot(time,Cai,color="blue",label="Baseline")
plt.plot(time_amylin,Cai_amylin,color="red",label="Amylin Incubation")
plt.plot(time_hip,Cai_hip,color="black",label="+HIP")
plt.legend(bbox_to_anchor=(1.75,1.0),fancybox=True, shadow=True)
plt.xlim([58.5,60])
plt.ylabel("Cai [mM]",**font)
plt.xlabel("Time [s]",**font)  
```

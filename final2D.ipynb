```python
job = job1
plt.plot(job.ts*dt,job.concsCaCleft,'r-',label="rapid/CaCleft")
plt.plot(job.ts*dt,job.concsCaSSL,'b-',label="rapid/CaSSL")
plt.plot(job.ts*dt,job.concsCa,'g-',label="rapid/Ca") 

job = job2
plt.plot(job.ts*dt,job.concsCaCleft,'r--',label="rapid/noSSL/CaCleft")
#plt.plot(job2.ts,job2.concsCaSSL,'b.',label="job2/CaSSL")
plt.plot(job.ts*dt,job.concsCa,'g--',label="rapid/noSSL/Ca") 
plt.legend()


job = job3
plt.plot(job.ts*dt,job.concsCaCleft,'r.',label="normal/CaCleft")
plt.plot(job.ts*dt,job.concsCaSSL,'b.',label="normal/CaSSL")
plt.plot(job.ts*dt,job.concsCa,'g.',label="normal/Ca") 


job = job4
plt.plot(job.ts*dt,job.concsCaCleft,'r.-',label="normal/noSSL/CaCleft")
#plt.plot(job.ts,job.concsCaSSL,'b.-',label="normal/noSSL/CaSSL")
plt.plot(job.ts*dt,job.concsCa,'g.-',label="normal/noSSL/Ca") 
plt.xlabel("time [ms]")
plt.ylabel("Ca [uM]")


plt.legend(bbox_to_anchor=[1.55,1])

plt.figure()
plt.plot(job1.ts*dt,job1.concsCa,'g-',label="Ca") 
plt.plot(job2.ts*dt,job2.concsCa,'g--',label="Ca") 
plt.plot(job3.ts*dt,job3.concsCa,'g.',label="Ca") 
plt.plot(job3.ts*dt,job3.concsCa,'g.-',label="Ca") 
plt.xlabel("time [ms]")
plt.ylabel("Ca [uM]")

#plt.legend(loc=0)
plt.gcf().savefig("test.png")
```

### Main routine for spawning off jobs for parameter analyses

Note that 'outputs' are defined by the user (outputList) and the operations to grab these outputs must be coded up

```python
numCores = 16  # maximum number of processors used at a time
numRandomDraws = 10 # number of random draws for parameters list in 'parmDict' (parmDict should probably be passed in)
jobDuration = 5e3  # [ms] simulation length 
dataFrame = fitter.ParameterSensitivity(numCores=numCores,numRandomDraws=numRandomDraws,jobDuration=jobDuration)
```

```python
ms_to_s = 1e-3

def test(dataSubChopped):       
    daMax = np.max(dataSubChopped)
    daMaxHalf = (daMax / 2)
    for i, val in enumerate(dataSubChopped):
            if daMax == val:
               daMaxPlace = i
            if val - daMaxHalf >= 0:
               daMaxHalfPlace = i
    result = (daMaxHalfPlace - daMaxPlace) * ms_to_s
    #result = dataSubChopped[daMaxPlace] - dataSubChopped[daMaxHalfPlace]
    print "daMaxPlace: ", daMaxPlace
    print "daMaxVal: ", dataSubChopped[daMaxPlace]
    print "daMaxHalfPlace: ", daMaxHalfPlace
    print "daMaxHalfVal: ", dataSubChopped[daMaxHalfPlace]
    return result
```

```python
def plotPandaPlot(rootOutput,df,outputs,inp=None,colors=None,xlabel=None,ylabel=None,
                   plot_xlim=None,plot_ylim=None,unit_scaler=None,legends=None,
                   legendMover1=None,legendMover2=None,legendPlacer=None):

    File_Name_Cases = ""
    xaxis_array = []
    std_input_array = []
    std_output_array = []
    std_output_over_input_array = []
    xaxis = df.as_matrix(columns=[inp])
    for j, vals in enumerate(xaxis):
        xaxis_array.append(vals[0])
    print "outputs: " , outputs
    for i,output in enumerate(outputs):
        print "output", output
        yaxis_array = []
        yaxis = df.as_matrix(columns=[output])
#        print yaxis
        yaxis *= unit_scaler[i]
        for k,vals in enumerate(yaxis):
#            print vals
            yaxis_array.append(vals[0])
#        print yaxis_array
        print "len x: ", len(xaxis_array)
#        print inp
        print "len y: ", len(yaxis_array)
#        df.plot(kind='scatter',x=inp,y=df.as_matrix(columns=[output]),color=colors[i],label=legends[i])
        plt.plot(xaxis_array,yaxis_array,"o",color=colors[i],label=legends[i])
        File_Name_Cases = "%s_" %legends[i]

        standardDeviationInput = np.std(xaxis_array)
        print "Std Input: ", standardDeviationInput
        std_input_array.append(standardDeviationInput)
        
        standardDeviationOutput = np.std(yaxis_array)
        print "Std Output: ", standardDeviationOutput
        std_output_array.append(standardDeviationOutput)
        
        standardDeviationOutputOverInput = standardDeviationOutput / standardDeviationInput
        print "Std Output over Input: ", standardDeviationOutputOverInput
        std_output_over_input_array.append(standardDeviationOutputOverInput)

    #plt.locator_params(nbins=8)
        plt.xlim(plot_xlim)
        plt.ylim(plot_ylim[i])

        plt.xlabel(xlabel, weight="bold",fontsize=14)
        plt.ylabel(ylabel[i], weight="bold",fontsize=14)
        plt.tight_layout()

        ax = plt.gca()
        ax.get_xaxis().get_major_formatter().set_useOffset(False)

        outFile = rootOutput+"Sensitivity_%s_%splots.png"%(inp,File_Name_Cases)
        print outFile

        art = []
        lgd = plt.legend(bbox_to_anchor=(legendMover1,legendMover2))
        art.append(lgd)

        plt.gcf().savefig(outFile,additional_artists=art,bbox_inches='tight',dpi=300)

        plt.show()
        plt.close()

    return std_input_array, std_output_array, std_output_over_input_array
```

```python
rootOutput="/net/share/shared/papers/despa/Brad/"
df_copy = df.copy()
#print df
inputs = ["Temp", "G_CaBk", "G_NaBk", "V_max_Jpump", "I_NaK_max", "PCa"]
outputs = ["Cai_amp", "Nai", "Ca_SR", "Cai_dia"]#, "APD", "Tau_Cai"]
colors = ['k','b','g','m','r','c']
xlabels = ["T [K]","G_CaBk [ms $\mathregular{uF^{-1}}$]","G_NaBk [ms $\mathregular{uF^{-1}}$]","V_max_Jpump [mM $\mathregular{ms^{-1}}$]",
           "I_NaK_max [uA $\mathregular{uF^{-1}}$]","PCa [$\mathregular{lF^{-1}}$ $\mathregular{ms^{-1}}$]"]
ylabels = ["Cai_amp [uM]","Nai [mM]","Ca_SR [uM]","Cai_dia [uM]"]#,"APD [s]","Tau [ms]"]
legends = ["Cai_amp","Nai","Ca_SR","Cai_dia","deltaBD","Tau"]
slicer = [df_copy.Temp, df_copy.G_CaBk, df_copy.G_NaBk, df_copy.V_max_Jpump, df_copy.I_NaK_max, df_copy.PCa]
slicerComparison = [310.0, 0.0007539, 0.001337, 0.000702, 5.0, 0.00111375]
plots_xlim = [[250,380],[6e-4,9e-4],[9e-4,1.8e-3],[6e-4,9e-4],[3.5,6.5],[9e-4,1.3e-3]]
unit_scalers = [[1.0e6],[1.0e3],[1.0e6],[1.0e6]]#,[1.0],[1.0]]
plots_ylim = [[-0.2,2.0],[11.0,11.5],[40,500],[0,6]]#,[5e-3,1.2e-2],[2,520]]
legendPlacer = None
legendMover1 = 1.3
legendMover2 = 1.0265

standardDevOutputOverInput_array =[]

for i,inp in enumerate(inputs):
    df_copy = df.copy()
    #print df_copy
    sliced_df = df_copy[slicer[i] != slicerComparison[i]]
    #print sliced_df
    standardDevInput, standardDevOutput, standardDevOutputOverInput = plotPandaPlot(rootOutput,sliced_df,outputs,inp,
                                                                                    colors,xlabels[i],ylabels,plots_xlim[i],
                                                                                    plots_ylim,unit_scalers,legends,legendMover1,
                                                                                    legendMover2,legendPlacer)
    print "stdInput: ", standardDevInput
    print "stdOutput: ", standardDevOutput
    print "stdOutputOverInput: ", standardDevOutputOverInput
    
    standardDevOutputOverInput_array.append(standardDevOutputOverInput)    

print "stdOutputOverInput_array: ", standardDevOutputOverInput_array
```

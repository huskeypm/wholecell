```python
class empty:pass


pacing = 1.0 # Hz
stim_period = 1000/pacing # [ms]
pacingSlow = 0.25 # Hz
stim_periodSlow = 1000/pacingSlow # [ms]

T = 1e5

caseDict = dict()

# baseline 
case = empty()
case.tag = "baseline1Hz"
case.label = "Baseline 1 Hz"

case.args = [
"-stim %d" % stim_period,
"-T %d" % T
]
case.name = "despa_%s_%d.pickle"%(case.tag,stim_period)
caseDict[case.tag] = case

# incr leak
case = empty()
case.tag = "2xincrleak1Hz"
case.label = "+100% Na/Ca leak 1Hz"
case.args = [
"-stim %d" % stim_period,
"-T %d" % T,
"-var G_CaBk 2. ",
"-var G_NaBk 2. ",
]
case.name = "despa_%s_%d.pickle"%(case.tag,stim_period)
caseDict[case.tag] = case


### SLOW 
# baseline 
case = empty()
case.tag = "baseline0.25Hz"
case.label = "Baseline 0.25 Hz"
case.args = [
"-stim %d" % stim_periodSlow,
"-T %d" % T
]
case.name = "despa_%s_%d.pickle"%(case.tag,stim_periodSlow)
caseDict[case.tag] = case

# incr leak
case = empty()
case.tag = "2xincrleak0.25Hz"
case.label = "+100% Na/Ca leak, 0.25 hZ"
case.args = [
"-stim %d" % stim_periodSlow,
"-T %d" % T,
"-var G_CaBk 2. ",
"-var G_NaBk 2. ",
]
case.name = "despa_%s_%d.pickle"%(case.tag,stim_periodSlow)
caseDict[case.tag] = case


### Delayed 
stim_start = 30 * 1e3
# baseline 
case = empty()
case.tag = "baseline1HzDelayed"
case.label = "Baseline 1Hz Delayed"
case.args = [
"-stim %d" % stim_period,
"-var stim_start %f " % stim_start,
"-T %d" % T
]
case.name = "despa_%s_%d.pickle"%(case.tag,stim_period)
caseDict[case.tag] = case

# incr leak
case = empty()
case.tag = "2xincrleak1HzDelayed"
case.label = "+100% Na/Ca leak, 1 Hz delayed"
case.args = [
"-stim %d" % stim_periodSlow,
"-T %d" % T,
"-var stim_start %f " % stim_start,
"-var G_CaBk 2. ",
"-var G_NaBk 2. ",
]
case.name = "despa_%s_%d.pickle"%(case.tag,stim_period)
caseDict[case.tag] = case


## Show all 
for key,case in caseDict.iteritems():
    print "# ", key
    print "python runShannonTest.py "+" ".join(case.args)+" -name "+case.name+" &"
#print "python runShannonTest.py "+" ".join(healthyArgs)+" -name "+healthyName+" &"
```

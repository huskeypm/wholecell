```python
def shiftNnorm(ctl_0p5Hz):
    minCtl_Cai_exp0p5Hz = np.amin(ctl_0p5Hz)
    Ctl_Cai_shift0p5Hz =  ctl_0p5Hz -  minCtl_Cai_exp0p5Hz
    maxCtl_Cai_exp0p5Hz = np.amax(Ctl_Cai_shift0p5Hz)
    Ctl_Cai_shift0p5Hz/=maxCtl_Cai_exp0p5Hz
    
    return Ctl_Cai_shift0p5Hz

def normShizzle(case0p5Hz,tRange = [int(290e3),int(291e3)]):
    s = case0p5Hz.data['s']
    s_idx = case0p5Hz.data['s_idx']
    idx = s_idx.index("Cai")
    t = case0p5Hz.data['t']

    cais = s[tRange[0]:tRange[1],idx]
    min0p5Hz_Cai = np.amin(cais)
    #print minWT0p5Hz_Cai
    max0p5Hz_Cai = np.amax(cais)
    max0p5Hz_Caix = np.argmax(cais)
    print max0p5Hz_Caix

    Comp_Cai_shift0p5Hz = cais - min0p5Hz_Cai
    #print Comp_Cai_shift
    maxComp_Cai_shift0p5Hz = np.amax(Comp_Cai_shift0p5Hz)
    Comp_Cai_shift0p5Hz/=maxComp_Cai_shift0p5Hz
    #print Comp_Cai_shift

    ms_to_s = 1e-3
    time=t[tRange[0]:tRange[1]]*ms_to_s
    
    return time, Comp_Cai_shift0p5Hz
```

```python

Ctl_Cai_shift0p5Hz = shiftNnorm(ctl_0p5Hz)
HIP_Cai_shift0p5Hz = shiftNnorm(hip_0p5Hz)
Ctl_Cai_shift2p0Hz = shiftNnorm(ctl_2p0Hz)
HIP_Cai_shift2p0Hz = shiftNnorm(hip_2p0Hz)

plt.plot(t_ctl_0p5Hz-0.12, Ctl_Cai_shift0p5Hz,"k",label="Control/0.5Hz, Despa 2012")
plt.plot(t_hip_0p5Hz-0.12, HIP_Cai_shift0p5Hz,"r",label="HIP/0.5Hz, Despa 2012")

plt.plot(t_ctl_2p0Hz, Ctl_Cai_shift2p0Hz,'--ok',label="Control/2.0Hz, Despa 2012")
plt.plot(t_hip_2p0Hz, HIP_Cai_shift2p0Hz,'--or',label="HIP/2.0Hz, Despa 2012")

#plt.xlim(289.8,292.4)
#plt.ylim(0.0,1.1)

plt.ylabel("Normalized $\mathregular{[Ca]_i}$",fontsize=16,weight="bold")
plt.xlabel("Time [S]",fontsize=16,weight="bold")
plt.title("Intercellular $\mathregular{Ca^{2+}}$ \n experimental 0.5Hz and 2.0Hz", fontsize=16,weight="bold")

art = []
lgd = plt.legend(bbox_to_anchor=(1.61,1.035))
art.append(lgd)

outFile=rootOutput+"Despa_0.5Hz_and_2.0Hz_Exp.png"
plt.tight_layout()
plt.gcf().savefig(outFile,additional_artists=art,bbox_inches='tight',dpi=300)
```

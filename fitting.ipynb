### Load SB model and integrate 
The ode model numerically solves for system states (s) that are described via ordinary differential equations, subject to time-independent parameters (pi):

ds/dt = p0 + p1*t 


'odeint' solves the ODE system ande returns a two-dimensional array representing the population of a given state (such as the intracellular Ca2+ concentration or the fraction of 'open' channels) as a function of time.  

In some cases, intermediate quantities will be used to update ds/dt:

<code>
j = p0*s0 - p1*s1
s0/dt = j
s1/dt = -j
</code>

These values can be computed ad hoc using the 'monitor' function.

## Fit fluxes for simple model 
Goal here is to fit some of the key calcium handling mechanisms to simple expressions that can be used with the PDE model. We start with the summation of all cleft/ssl currents, which is represented by the 'J_Ca_SL_myo' entry (defined in the shannon_2004.ode):

<code>J_Ca_SL_myo = (Ca_SL - Cai) x 3.7243e-12 # mmole*ms**-1</code>


Note that although the J_Ca term represents the flux due to diffusion between compartments, Ca_SL reflects the calcium concentration as a result of all cleft/SSL fluxes working (e.g. RyR, LCC, NCX, etc). Hence, we assume that the J_Ca reflects the flux of the major calcium entry mechanisms.

Comparison of RyR release with diffusional flux from SL compartment. I'm surprised that the RyR release is so large, but the predicted values agree with the opencell results. Fig 4B in the Shannon Bers paper reports a maximum amplitude of 3.5 [mM/s] for SR Ca release, although the SB model predicts more than 10 times this (52 [uM/ms] <--> [mM/s]. However, the expression for SR release is 

$ j = O (Ca_{SR}-Ca_i) ks = O (\Delta Ca) ks $

for which $max(O) \approx  10\cdot10^{-3}$ (agrees with Fig 4D), $max(\Delta Ca) < 1. $ [mM], $ks = 25$ [1/ms], so $j < 0.25$ [mM/ms] / 250 [uM/ms]. Based on the default model parameters, the predicted maximum pOpen (1.2e-2,above) and Delta Ca (0.35), our max is closer to 100 [uM/ms], as shown in the plots below (blue).  

However, for the junction-> SL flux we have a maximum at 0.9 uM/ms; this is a bit short of the SR Ca release profile in Fig 92 (pg 177) of the Bers book (~25 mM/s --> 2.5 [uM/ms] from Shannon 2000).

#### Grab representative segment of transients for doing exponential fitting  
Steady-state appears around t= 5000

```python
#A = 9.5 # pA
#tau = 2 # ms 
waveformStart = 5000 # ms 
waveformEnd = waveformStart + stim_period # ms 
above = np.where(tsteps>waveformStart)
below = np.where(tsteps<waveformEnd)


z = np.in1d(above,below)
inds = above[0][z]

inds2 = inds[np.where(tsteps[inds]>waveformStart)]
tsub = tsteps[inds2]
js =jsJ_Ca_SL_myo/volMyo # divide by volMyo as in .ode code 
jsub =js[inds2] # 
plt.plot(tsub,mM_to_uM*jsub,'b')
plt.xlabel("t [ms]")
plt.ylabel("j [uM/ms]")
```

We'll assume that the current instantaneously activates at t for which j(t) is maximal, after which it follows an exponential decay. The function below ensures that we 'grab' a representative interval, finds its max, etc

```python
import taufitting as tf
case={'t':tsteps,'s':js} # hijacking the function, which is normally intended for decay of Ca2+ transients

tsub,jsub = tf.GetInterval(case, stim_period , tstart=waveformStart,idx=False,doPlot=True)
```

We now fit the following expression for the flux

$f = A\; e^{(-t/\tau)} $

A is determined by the maximum current (e.g. at t=0). Note that my rule of thumb (see my lecture notes) is that $1 [A/F] = 0.02 [\mu M/ms]$ for rabbit ventricular myocytes. 

Tau is obtained by nonlinear fitting via taufitting.py


Below we obtain A=44 [A/F]. This is probably reasonable (although still a bit short), given that Fig 5B in Shannon 2004 suggests LCC current is about 20 [A/F], and SR current is at least twice this. tau also looks plausible based on figure 4B of Shannon 2004. 

```python
A_uMms = np.max(jsub)* mM_to_uM # max flux [uM/ms]
pA_to_uMms = 2e-2
A= A_uMms / pA_to_uMms # A [uM/ms] --> [A/F]

tau = tf.GetTau(case,stim_period,waveformStart,False) # time-decay [ms] 

print "A %f [A/F] tau %f [ms] " % (A, tau)


```

```python
def SERCAExpression(cai=0.1):
  # CellML values
  Vmax= 5.311e-3*mM_to_uM # uM/ms
  Kmf = 0.246 # uMs
  Kmr = 1.7e3 # mu (note: Shannon uses 1.7 mM, which I believe is for total Ca. However, free SR Ca seems to track exactly with total Ca)
  H = 1.787
  Q10 = 2.6
    
  # Grabbing cellML stuff   
  #T = 310
  #casr=500  
  #Ca_SR = casr  
  #Cai = cai   
  #H_Jpump = H  
  Q_SRCaP = 1 # math.pow(Q10, -31 + T/10.)
  casrTerm = 0.112265722393 #  math.pow(Ca_SR/Kmr, H_Jpump)  --> 0.112265722393 

 # print casrTerm
  j_pump_SR = Vmax*(-casrTerm +\
        math.pow(cai/Kmf, H))*Q_SRCaP/(1 + casrTerm\
        + math.pow(cai/Kmf, H))
  ################################
    
  return j_pump_SR

def SERCAShannonExpression(cai=0.1, casr=500):
  # SHannon paper values   
  Vmax=286. # uM/ms
  Kmf = 0.246 # uM
  Kmr = 1.7e3 # mu (note: Shannon uses 1.7 mM, which I believe is for total Ca. However, free SR Ca seems to track exactly with total Ca)
  H = 1.787
  Q10 = 2.6

    
  # CellML values
  Vmax= 5.311e-3*mM_to_uM # uM/ms
  Kmf = 0.246 # uM
  Kmr = 1.7e3 # mu (note: Shannon uses 1.7 mM, which I believe is for total Ca. However, free SR Ca seems to track exactly with total Ca)
  H = 1.787
  Q10 = 2.6
    
  ## PKH params 
  ##Vmax=320. # uM/s (by setting Vmax to zero and rescaling so that maxcai/mincasr gives J=205 uM/s)
  ##Vmax = 300
  #Kmf = 0.100 # uM # if set to the normal value, the pump goes in reverse
  #Kmr = 0.55e3 # uM (this is my adjusted value based on free Ca)
  #H = 1.0
  #Q10=1 # PKH

  # Grabbing cellML stuff   
  T = 310
  V_max_Jpump = Vmax  
  Q_SRCaP = math.pow(Q10, -31 + T/10.)
  Ca_SR = casr  
  Cai = cai   
  H_Jpump = H  
  j_pump_SR = V_max_Jpump*(-math.pow(Ca_SR/Kmr, H_Jpump) +\
        math.pow(Cai/Kmf, H_Jpump))*Q_SRCaP/(1 + math.pow(Ca_SR/Kmr, H_Jpump)\
        + math.pow(Cai/Kmf, H_Jpump))

  # from cellML  
  #print Cai,Ca_SR,T,j_pump_SR , V_max_Jpump,(-math.pow(Ca_SR/Kmr, H_Jpump) ,
  #      math.pow(Cai/Kmf, H_Jpump)),Q_SRCaP/(1 + math.pow(Ca_SR/Kmr, H_Jpump)\
  #      + math.pow(Cai/Kmf, H_Jpump))
    
    
  return j_pump_SR

# debug and asserts
# Note: SB model outputs are in mM, not uM as I am using in the function above 
#0.000500188262832 0.343930077613 310.0 0.00402719730581 0.0053114 (-0.057525440018213296, 3.5542708683270097) 0.216835248814
cai = 0.000500188262832*mM_to_uM; casr= 0.343930077613*mM_to_uM; j_pump_SR = 0.00402719730581*mM_to_uM
assert( np.abs(SERCAShannonExpression(cai=cai, casr=casr)  - j_pump_SR) < 1e-3)

# 0.000101898188168 0.576176631447 310.0 0.000245072273418 0.0053114 (-0.1446440943254374, 0.20701052708214052) 0.739833966578
cai = 0.000101898188168*mM_to_uM; casr= 0.576176631447*mM_to_uM;  j_pump_SR = 0.000245072273418*mM_to_uM
assert( np.abs(SERCAShannonExpression(cai=cai, casr=casr)  - j_pump_SR) < 1e-3)

vfunc1 = np.vectorize(SERCAExpression)    
vfunc2 = np.vectorize(SERCAShannonExpression)    
plt.plot(tsub,vfunc1(cais), label="Simplified Shannon model")
plt.plot(tsub,vfunc2(cais), label="Shannon model")
plt.ylabel("j [uM/ms]")
plt.legend()
```

It's not possible to find a single best fit buffer for the shannon-bers data (clear bifurcation )
Most likely it's because the kinetics are too slow. 

```python
## Function defs
# Function for generating data begin
def GenData(vars,c):
    bmax = vars[0]
    kd = vars[1]
    #cb = x 
    #c = -(cb*kd)/(cb-bmax)
    cb = bmax/(kd/c + 1)
    return cb

# residual (for optimization)
eps_data = 0.1
def residual(vars, x, data, eps_data):
    model = GenData(vars,x)
    return (data-model)/eps_data


## Generate data
mode = "real"
#mode = "test"
if mode=="test":
  vars = [70., 1.]
  cais = np.linspace(0.1,1,100)
  cbtots = GenData(vars,cais) + 1.*np.random.randn(np.shape(cais)[0])

if mode=="real":
  vars = [70., 1.]    
  cais = sbCai[inds]
  cbtots= sbCbTot[inds]


x = cais
data = cbtots

## fitting 
from scipy.optimize import leastsq
out = leastsq(residual, vars, args=(x, data, eps_data))
outvars = out[0]
print "Outvars ", outvars

## Compare
dataFit = GenData(outvars,x)


plt.plot(x*mM_to_uM,data,label="data")
plt.plot(x*mM_to_uM,dataFit,label="Fit")
plt.xlabel("cai")
plt.ylabel("CaB")
plt.legend(loc=0)
```
